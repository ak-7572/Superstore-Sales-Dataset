```python
# checking if their are any null values
df.isna().sum()
```

```python
# filling the null values (we can even drop this coulumn if that coulumn is not important)
df['Postal Code'].fillna(0, inplace = True)
```

```python
# rechecking for null values
df.isna().sum()
```

```python
# converting order date and ship date to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst = True)
df['Ship Date'] = pd.to_datetime(df['Ship Date'], dayfirst = True)
```

```python
# creating a shipping dealy to extract the delay days.
df['shipping Delay Days'] = df['Ship Date'] - df['Order Date']
df['shipping Delay Days'] = df['shipping Delay Days'].dt.days
```

```python
df['shipping Delay Days']
```

```python
df.drop('shipping Delay', axis = 1, inplace = True)
```

```python
# shipping dealy coulumn added in the dataset
df.head(3)
```

```python
sns.histplot(df['shipping Delay Days'], bins = 10, kde = True)
plt.title('Distribution of Shipping Delay Days')
plt.xlabel('Shipping Delay Days')
plt.ylabel('No of orders')
plt.show()
```

```python
# from the above chart we can see that :-
# the max days in delivering the order is 7 days
# the most shipping delay is 4 days, where the no of orders has crossed 2500
# Therefore, the company must focus on reducing the the delivery days of 7 days as the cutomer might get upset and angry and at the end moment they might cancel the oder.
```

```python
top_customers = df.groupby('Customer Name')['Order ID'].count().sort_values(ascending = False).head(5)
top_customers
```

```python
top_customers.plot.bar()
plt.title('top 5 customers details')
plt.xlabel('Customer Name')
plt.ylabel('No of orders')
plt.show()
```

```python
# from above data we can see the companies top 5 customers with the no of orders.
# William Brown is the first customer who has order 35 times.
# Seth Vernon is the 5th customer with 32 no of orders.
```

```python
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending = False)
category_sales
```

```python
# total sales by category
sns.barplot(x = category_sales.index, y = category_sales.values)
plt.title('Category wise sales')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

```python
# from the above data we can see that the Technology items has been sold more with more than Rs.8L
# 2nd most sold item is Furniture with more than Rs.7L
# 3rd most sold item is office supply with more than Rs.7L
```

```python
# sales by region
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending = False)
sns.barplot(x = region_sales.index, y = region_sales.values)
plt.title('region wise sales')
plt.xlabel('region')
plt.ylabel('Sales')
plt.show()
```

```python
# the above char shows that west side has ordered more products
# 2nd most products are ordered from east region & 3rd is central region
# 4th most products are ordered from south region
```

```python
# monthly sales trend

df['order date'] = pd.to_datetime(df['Order Date'])
df['month'] = df['order date'].dt.to_period('M')

monthly_sales = df.groupby('month')['Sales'].sum()

monthly_sales.plot(kind = 'line', marker = 'o', figsize = (12,5))
plt.title('Monthly Sales Trend'),
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
# top 5 products by sales

top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending = False).head(5)

sns.barplot(x = top_products.index, y = top_products.values)
plt.title('Top 5 Products by Sales')
plt.xlabel('Product Name')
plt.ylabel('Sales')
plt.xticks(rotation = 90)
plt.show()
```
